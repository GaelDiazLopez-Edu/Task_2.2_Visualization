# Time Series (Evolution) Plots

**What do they represent?**

Evolution of data over time

**What are they useful for?**
- **Trends**: Is the value generally going up or down over a long period?
- **Seasonality**: Are there patterns that repeat at fixed intervals (e.g., higher sales every weekend or every December)?
- **Noise**: Random variations that don't follow a pattern.
- **Outliers**: Spikes that don't match the history (like a sudden crash in stock prices).


## Trends

###  Trend sample

The "Keeling Curve" (CO2 Concentration)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Dataset: Monthly CO2 concentrations at Mauna Loa Observatory
url_co2 = "https://raw.githubusercontent.com/datasets/co2-fossil-global/master/global.csv"
df_co2 = pd.read_csv(url_co2)

# Visualization
plt.figure(figsize=(10, 5))
plt.plot(df_co2['Year'], df_co2['Total'], color='black', linewidth=2)
plt.title('Real Data: Fossil Fuel CO2 Emissions (Global Trend)')
plt.xlabel('Year')
plt.ylabel('Million Metric Tons of Carbon')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

 # 1. Life Expectancy (Global Health Trend)

 ## 1.1 Plot the life expectancy in 'Spain'

 Write an interpretation of the graph below it in Markdown format.

```python
# We'll use a curated World Bank subset (Life Expectancy at birth, World average)
url_life = "https://raw.githubusercontent.com/datasets/gdp/master/data/gdp.csv"
```

```python
# 1. Cargar los datos desde la URL del notebook
df = pd.read_csv(url_life)

# 2. Filtrar para obtener solo los datos de Espaa
# Usamos 'Country Name' en este dataset
df_spain = df[df['Country Name'] == 'Spain']

# 3. Crear el grfico de evolucin
plt.figure(figsize=(10, 5))
plt.plot(df_spain['Year'], df_spain['Value'], color='blue', marker='o', markersize=4)

# 4. Personalizacin tcnica
plt.title('Evolucin del PIB en Espaa') # Pone esperanza de vida en el notebook pero interpreto que es PIB ya que el numero es muy grande
plt.xlabel('Ao')
plt.ylabel('USD')
plt.grid(True, alpha=0.3)

plt.show()
```

## 1.2 Plot the life expectancy in 'Spain' vs 'France'

Write an interpretation of the graph below it in Markdown format.

```python
# 1. Cargar los datos (usando la variable url_life ya definida en el notebook)
df = pd.read_csv(url_life)

# 2. Filtrar los datos para Espaa y Francia
df_spain = df[df['Country Name'] == 'Spain']
df_france = df[df['Country Name'] == 'France']

# 3. Crear el grfico comparativo
plt.figure(figsize=(12, 6))

# Dibujamos ambas lneas
plt.plot(df_spain['Year'], df_spain['Value'], label='Espaa', color='red', linewidth=2)
plt.plot(df_france['Year'], df_france['Value'], label='Francia', color='blue', linewidth=2)

# 4. Personalizacin del grfico
plt.title('Comparativa de Evolucin Econmica: Espaa vs Francia')
plt.xlabel('Ao')
plt.ylabel('USD')
plt.legend() # Para saber qu lnea es cada pas (El cuadrito de arriba a la izquierda)
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

Como podemos ver en el grafico ambas economicas tienden a crecer aun que veamos que sufren caidas en las diferentes crisis. La economica francesa desde el 1987 aprox. empez a crecer mas rapido que la espaola.

## 1.3 Plot the life expectancy in 'Spain' vs 'France' vs 'China'

Write an interpretation of the graph below it in Markdown format.

```python
# 1. Cargar los datos (usando la variable url_life definida anteriormente)
df = pd.read_csv(url_life)

# 2. Filtrar los datos para los tres pases
df_spain = df[df['Country Name'] == 'Spain']
df_france = df[df['Country Name'] == 'France']
df_china = df[df['Country Name'] == 'China']

# 3. Crear el grfico comparativo
plt.figure(figsize=(12, 6))

# Dibujamos las tres lneas
plt.plot(df_spain['Year'], df_spain['Value'], label='Espaa', color='red')
plt.plot(df_france['Year'], df_france['Value'], label='Francia', color='blue')
plt.plot(df_china['Year'], df_china['Value'], label='China', color='green', linewidth=3)

# 4. Personalizacin del grfico
plt.title('Comparativa Econmica Global: Espaa vs Francia vs China')
plt.xlabel('Ao')
plt.ylabel('USD (Escalado arriba izquierda)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

Como podemos observar la economia China a partir de 2005 aument muchas ms que las de Espaa y Francia. Aun que todas tiendan a crecer. El escalado lo tenemos en la parte superior izquierda para saber verdaderamente los numeros en USD de la economa.

 Write an interpretation of the graph below it in Markdown format.

```python
url_spotify = 'https://raw.githubusercontent.com/bencmbit/datasets/refs/heads/master/spotify.csv'
```

```python
# 1. Cargar el dataset de Spotify
# Usamos index_col='Date' y parse_dates=True para que Python reconozca el tiempo automticamente
df_spotify = pd.read_csv(url_spotify, index_col='Date', parse_dates=True)

# 2. Crear el Multiplot
# Pandas permite hacer esto muy fcil con el argumento subplots=True
df_spotify.plot(subplots=True, figsize=(10, 12), sharex=True)

# 3. Personalizacin
plt.suptitle('Evolucin individual de xitos en Spotify', fontsize=16)
plt.xlabel('Fecha')
plt.show()
```

 # 3. Seasonality in Airline Passengers (1949-1960)

 ## 3.1 Draw lineplots to analize Seasonality (and Trend)

 Write an interpretation of the graphs below it in Markdown format.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Cargar el dataset
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
try:
    df = pd.read_csv(url)
except:
    df = sns.load_dataset('flights')
    df.columns = ['Year', 'Month', 'Passengers']
    # Crear columna de fecha para consistencia
    df['Month'] = pd.to_datetime(df['Year'].astype(str) + '-' + df['Month'].astype(str))
    df = df[['Month', 'Passengers']]

# 2. Preparar los datos
df['Month'] = pd.to_datetime(df['Month'])
df.set_index('Month', inplace=True)
df['Year'] = df.index.year
df['Month_Num'] = df.index.month

# 3. Grfico de Tendencia General
plt.figure(figsize=(12, 5))
sns.lineplot(data=df, x=df.index, y='Passengers')
plt.title('Pasajeros Mensuales de Aerolneas (1949-1960) - Tendencia Global')
plt.xlabel('Ao')
plt.ylabel('Miles de Pasajeros')
plt.grid(True)
plt.show()

# 4. Grfico de Estacionalidad (Lneas por ao)
plt.figure(figsize=(12, 5))
sns.lineplot(data=df, x='Month_Num', y='Passengers', hue='Year', palette='viridis')
plt.title('Estacionalidad: Comparativa Mensual por Ao')
plt.xlabel('Mes (1=Enero, 12=Diciembre)')
plt.ylabel('Pasajeros')
plt.xticks(range(1, 13))
plt.legend(title='Ao', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Podemos ver una gran tendencia ascendente a largo plazo. La industria area en esa decada vemos que funcion muy bien, puede ser por que fue bajando el precio y ms clientes se lo podian permitir. 
Vemos un pico muy grande en julio y agosto que coinciden con las vacaciones de los estudiantes y donde muchos trabajadores suelen intentar coger. 
Tambien vemos que en esas fechas cada ao va aumentando aun mas la gente que coge vuelos indicando los viajes por estaciones aumentan segn sube la tendencia.
